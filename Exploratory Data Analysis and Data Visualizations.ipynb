```python
#Importing required libraries

import pandas as pd
import matplotlib as plt
import seaborn as sns

%matplotlib inline
```

```python
#Loading data

train = pd.read_csv(r'C:\Users\SKothapally\Downloads\Walmart\train.csv')
features = pd.read_csv(r'C:\Users\SKothapally\Downloads\Walmart\features.csv')
stores = pd.read_csv(r'C:\Users\SKothapally\Downloads\Walmart\stores.csv')
```

# Bi Variate PLots

```python
# Plots showing weekly sales of each store

fig = plt.figure(figsize = (10,10))
ax = fig.gca()
sns.set_style('whitegrid')
sns.barplot(x = merged_data['Store'], y = merged_data['Weekly_Sales'],ax = ax)
ax.set_title('Weekly Sales of each store')
ax.set_xlabel('Store')
ax.set_ylabel('Weekly_Sales')
```

From the above plot we see for stores (1,2,4,10,13,14,19,20,23,31,39) have good weekly sales and others stores are under performing. Among the under performing stores (3,5,7,9,15,16,29,30,33,36,37,38,44) are poorly performing stores. Exploring further might help us understand the reason for underperforming stores.

```python
#Plot of Temperature vs Weekly_sales

fig = plt.figure(figsize = (10,10))
sns.regplot(x = merged_data['Temperature'], y = merged_data['Weekly_Sales'], fit_reg = False)
ax.set_title('Temperature vs Weekly Sales')
ax.set_xlabel('Temperature')
ax.set_ylabel('Weekly_Sales')
```

Days when temperature falls down to 0 degrees or goes higher to 100 degrees have low sales as customers tend to stay home on extreme climatic conditions. Good sales are seen when temperatures are in the range of 40 degrees to 70 degrees.

```python
#Plot of Fuel Price vs Weekly_sales

fig = plt.figure(figsize = (10,10))
ax = fig.gca()
sns.set_style('whitegrid')
sns.regplot(x = merged_data['Fuel_Price'], y = merged_data['Weekly_Sales'],ax = ax, fit_reg = False)
ax.set_title('Fuel Price vs Weekly Sales')
ax.set_xlabel('Fuel Price')
ax.set_ylabel('Weekly_Sales')
```

Fuel price has no influence on sales.
